# Using inside Jupyter notebook

Load th cythonmagic extension.

````python
%load_ext cython
````

Then, simply use the magic function to start writing cython code.

````python
%%cython

cdef int a = 0
for i in range(10):
    a += i
print(a)
````

Add `--annotate` or `-a` for showing a code analysis of the compiled code

````python
%%cython --annotate

cdef int a = 0
for i in range(10):
    a += i
print(a)
````
